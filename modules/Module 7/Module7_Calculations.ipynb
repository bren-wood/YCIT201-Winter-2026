```python

import pandas as pd
import statsmodels.api as sm

# Brokerage data (exactly as in Excel)
data = {
    "Price": [3.4,3.2,3.1,2.9,2.9,2.5,2.6,2.4,2.6,2.3,3.7,2.5,3.0,4.0],
    "Speed": [3.4,3.3,3.4,3.6,3.2,3.2,3.8,3.8,2.6,2.7,4.0,2.5,3.0,1.0],
    "Satisfaction": [3.5,3.4,3.9,3.7,2.9,2.7,2.8,3.6,2.6,2.3,4.0,2.5,4.0,2.0]
}

df = pd.DataFrame(data)

# OLS regression
X = df[["Price", "Speed"]]
X = sm.add_constant(X)
y = df["Satisfaction"]

model = sm.OLS(y, X).fit()
print(model.summary())
```

```python

# New observation (same as Excel)
x_new = pd.DataFrame({
    "const": [1],
    "Price": [3],
    "Speed": [3]
})

prediction = model.get_prediction(x_new)
results = prediction.summary_frame(alpha=0.05)

print(results)
```
